```python
# Genero un dataframe con las columnas que necesito
base_red = base[["Beneficio","Parentesco","Apellido y Nombre","CodDiagnostico","Cod Prestación","Fecha Práctica"]].copy()

# Cambio el tipo de dato fecha
base_red["Fecha Práctica"] = pandas.to_datetime(base_red["Fecha Práctica"])

# Ordeno los datos
base_red.sort_values(by=['Beneficio','Parentesco','CodDiagnostico','Fecha Práctica'], inplace=True)
```
